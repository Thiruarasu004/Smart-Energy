# feature_selection

```python
sns.heatmap(df[['energy_consumption', 'building_id', 'apparent_temperature',
       'air_temperature', 'dew_point_temperature', 'relative_humidity',
       'wind_speed', 'wind_direction', 'month', 'day', 'hour',
       'weekday', 'is_holiday','building_category',"building_builtyear" ,'building_floorarea', 'is_weekend',
       'season']].corr(), annot=True, fmt=".2f", cmap="YlGnBu", annot_kws={"size":7 })
```

```python
x=df.drop(["energy_consumption","timestamp"],axis=1)
y=df[["energy_consumption"]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=0)
```

```python
rf_regressor.fit(x_train, y_train)
```

```python
y_predict=rf_regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_predict,y_test)
```

# Extra Trees Regressor

```python
from sklearn.ensemble import ExtraTreesRegressor
```

```python
extra_trees_regressor = ExtraTreesRegressor(n_estimators=100, random_state=0)
extra_trees_regressor.fit(x_train, y_train)
```

```python
y_pred = extra_trees_regressor.predict(x_test)
```

```python
y_train_pred = extra_trees_regressor.predict(x_train)
```

```python
r2_score(y_pred,y_test)
```

```python
r2_score(y_train_pred,y_train)
```

## Light Gradient Boosting

```python
import lightgbm as lgb
```

```python
train_data = lgb.Dataset(x_train, label=y_train)
test_data = lgb.Dataset(x_test, label=y_test, reference=train_data)


params = {
    'objective': 'regression',
    'metric': 'l2',
    'boosting_type': 'gbdt',
    'num_leaves': 31,
    'learning_rate': 0.05,
    'feature_fraction': 0.9
}
num_round = 100
bst = lgb.train(params, train_data, num_round, valid_sets=[test_data])
lgb_y_pred = bst.predict(x_test, num_iteration=bst.best_iteration)
```

```python
r2_score(lgb_y_pred,y_test)
```

```python
sns.set(style="whitegrid")
g = sns.FacetGrid(tested_df, col="building_id", col_wrap=3, height=4,sharey=False)

g.map(sns.lineplot, "timestamp", "energy_consumption")
g.map_dataframe(lambda data, color: sns.scatterplot(data=data[data["if_anoalies"] == -1], x="timestamp", y="energy_consumption", color="red", marker="o"))

g.set_titles("Building {col_name}")
g.set_axis_labels("Timestamp", "Electricity Consumption")
g.set_xticklabels(rotation=90)


plt.tight_layout()
plt.show()
```
