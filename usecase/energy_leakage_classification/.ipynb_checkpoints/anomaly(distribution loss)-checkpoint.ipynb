# leakage classification detection

## bulk1

```python
bulk1["sum_of_consumption"]=bulk1[8.0]+bulk1[9.0]
```

```python
bulk1["distribution_loss"]=bulk1["consumption"]-bulk1["sum_of_consumption"]
```

```python
bulk1.loc[bulk1['distribution_loss'] < 0, 'distribution_loss'] = 0
```
